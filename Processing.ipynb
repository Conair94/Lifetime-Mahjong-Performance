```python
df_data = pd.read_csv("data.csv")
names = ['connor','helio','sam','oscar','nick','drew','ruohan','brandon','ruohan']
df = df_data[names].cumsum()
#Convert to Dollars
df = df/100

print(df.head(5))
#df.to_csv("results.csv", index = False)
```

Creating Plots

```python
df.plot()

plt.xlabel("Game Number")
plt.ylabel("Net Profit")
plt.legend()
plt.savefig("net_profit.png")
```
